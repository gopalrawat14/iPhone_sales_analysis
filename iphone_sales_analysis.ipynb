# iPhone Sales Analysis

This project focuses on analyzing sales data for iPhones, aiming to uncover patterns and insights related to pricing, customer preferences, and product performance. The dataset includes comprehensive details about various iPhone models and their market reception.

## Key Features of the Dataset

Product Name: Names of iPhone models.
Product URL: Links to product pages for reference.
Sale Price, MRP, and Discounted Price: Information about pricing and discounts.
Customer Feedback:
Number of Ratings: Total customer ratings.
Number of Reviews: Count of textual reviews.
Star Rating: Average star rating given by customers.
Technical Specifications:
RAM: Memory configurations for each model.
UPC: Universal Product Code for identification.
## Objectives

Identify pricing trends and the effectiveness of discounts.
Analyze customer feedback to gauge satisfaction and preferences.
Determine the impact of technical specifications on sales performance.
Provide actionable insights for marketing and product strategy.

# Question1: What are the top 10 highest-rated iPhones on Flipkart in India?

```python
data.head()
```

```python
high_rated= data.sort_values(by=["Star Rating"], ascending=False) #we get rating in descending order after ascending=False
high_rated= high_rated.head(10)
print(high_rated['Product Name'])
```

```python
print(high_rated)
```

# Question2: How many ratings do the highest-rated iPhones on Flipkart have?

```python
data.head()
```

```python
#bar graph
iphones= high_rated['Product Name'].value_counts()
label= iphones.index
counts= high_rated["Number Of Ratings"]
fig= px.bar(high_rated, x=label, y=counts, title= "Number of ratings of highest rated iphones")
fig.show()
```

# Conclusion: The highest rated iphone 8 Plus(Gold, 64 GB) got 3431 ratings in total.

# Question3: Which iPhone has the highest number of reviews on Flipkart?

```python
data.head()
```

```python
#bar graph
iphones= high_rated['Product Name'].value_counts()
label= iphones.index
counts= high_rated["Number Of Reviews"]
fig= px.bar(high_rated, x=label, y=counts, title= "Number of reviews of highest rated iphones")
fig.show()
```

# Conclusion: The iphone 8 Plus(Gold,64GB) has the highest number of reviews(356) on flipkart.

# Question4: What is the relationship between the sale price of iPhones and the number of ratings on Flipkart?

```python
data.head()
```

```python
#scatter plot
figure = px.scatter(data_frame= data,
                    x="Number Of Ratings",
                    y="Sale Price",
                    size="Discount Percentage",
                    trendline="ols",
                    title="Relationship between Sale Price and Number of Ratings of iphones.")
figure.show()
```

# Conclusion: There is a negative linear relationship between the sale price and the number of ratings. i.e., iphones with lower sale price are being soled more.

# Question5: What is the relationship between the discount percentage and the number of ratings of iPhones on Flipkart?


```python
data.head()
```

```python
#scatter plot
figure=px.scatter(data_frame=data,
                  x="Number Of Ratings",
                  y="Discount Percentage",
                  size="Sale Price",
                  trendline="ols",
                  title="Relationship bw Discount Price and Number Of Ratings")

figure.show()
```

# Conclusion: There is a linear relationship between Discount Price and Number of Ratings.i.e., people buys those iphones which are having more discounted price.

# Question6: Can you figure out the least expensive and most expensive iPhones in the Indian market, alone with all their specifications?

```python
data.head()
```

```python
most_expensive= data.loc[data['Sale Price'].idxmax()]
least_expensive= data.loc[data['Sale Price'].idxmin()]
print("Most Expensive Iphone is: ", most_expensive)

print("\nLeast Expensive Iphone is: ",least_expensive)

```

# Conclusion: The most expensive iphone is APPLE iPhone 12 Pro (Silver, 512 GB) and the least expensive iphone is APPLE iPhone SE (White, 64 GB) .
